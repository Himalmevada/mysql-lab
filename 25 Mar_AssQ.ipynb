```python
plt.boxplot(x=df["Price"])
plt.show()

Q1 = np.percentile(df["Price"], 25)
Q3 = np.percentile(df["Price"], 75)
IQR = Q3 - Q1

lower_whisker = Q1 - 1.25 * IQR
upper_whisker = Q3 + 1.25 * IQR

print(lower_whisker,upper_whisker)

# The value above 21243 are considered as outlier in the price feature.
```

## Outliers can significantly impact your analysis and statistical models in various ways:

- Influence on Measures of Central Tendency: Outliers can skew the mean (average) and make it less representative of the typical data. The median is less affected by outliers, making it a more robust measure of central tendency.

- Effect on Variability: Outliers can inflate the standard deviation and other measures of variability, making the data appear more spread out than it actually is.

- Model Performance: In regression and other predictive models, outliers can have a disproportionate influence on the model's fit and predictions, leading to biased results.

- Assumptions Violation: Many statistical techniques assume that data follow a normal distribution or have constant variance. Outliers can violate these assumptions, affecting the validity of the analysis.

- Reduced Generalization: Outliers in training data may not be representative of the underlying population, leading to poor generalization of the model to new data.
