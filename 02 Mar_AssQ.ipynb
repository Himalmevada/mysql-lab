### Ans - 1

#### Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.
#### Like of if give you 10lac rows data in excel. It will really hard to understand and take the insights from that data.
#### For that we need to make it easy to understandable for everyone. So we apply visual chart to make more understandable.
#### Once we got the charts we can easily found the pattern or relations between different entity or features.

#### The six most commonly used Plots come under Matplotlib. These are:
1. Line Plot
2. Bar Plot
3. Scatter Plot
4. Pie Plot
5. Histogram Plot

#### **Line plots** are drawn by joining straight lines connecting data points where the x-axis and y-axis values intersect. Line plots are the simplest form of representing data.
#### **The bar plots** are vertical/horizontal rectangular graphs that show data comparison where you can gauge the changes over a period represented in another axis (mostly the X-axis). Each bar can store the value of one or multiple data divided in a ratio. The longer a bar becomes, the greater the value it holds.
#### We can implement **the scatter (previously called XY) plots** while comparing various data variables to determine the connection between dependent and independent variables. The data gets expressed as a collection of points clustered together meaningfully. Here each value has one variable (x) determining the relationship with the other (Y).
#### **A pie plot** is a circular graph where the data get represented within that components/segments or slices of pie. Data analysts use them while representing the percentage or proportional data in which each pie slice represents an item or data classification.
#### We can use a **histogram plot** when the data remains distributed, whereas we can use a bar graph to compare two entities. Both histogram and bar plot look alike but are used in different scenarios.


### Ans - 2

#### A scatter chart, also called a scatter plot, is a chart that shows the relationship between two variables. They are an incredibly powerful chart type, allowing viewers to immediately understand a relationship or trend, which would be impossible to see in almost any other form.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x=x,y=y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("MY SCATTER PLOT")
```

### Ans - 3

#### Sometimes it is helpful to compare different views of data side by side. To this end, Matplotlib has the concept of subplots: groups of smaller axes that can exist together within a single figure. These subplots might be insets, grids of plots, or other more complicated layouts.

```python
# line_1: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# line_2: 
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# line_3: 
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# line_4: 
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(15,5))
plt.subplot(2,3,1)
plt.plot(x1,y1)

plt.subplot(2,3,2)
plt.plot(x2,y2)

plt.subplot(2,3,3)
plt.plot(x3,y3)

plt.subplot(2,3,4)
plt.plot(x4,y4)

plt.show()
```

### Ans - 5

#### A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```
