### Que - 1 : Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
from scipy.stats import norm

sample_mean = 50
sample_std = 5
ci = 0.95
lower, upper = norm.interval(confidence=ci,loc=sample_mean,scale=sample_std)

print(lower,upper)
```

### Que - 2 : Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np
import scipy.stats as stat

observed = np.array([18,22,19,11,15,15])
expected_data = np.array([20,20,20,10,10,20])
expected = [expected_data * 100 / 100]
significance_val = 0.05
ci = 1 - significance_val

chi_square = np.sum((observed - expected)**2 / expected)
print(chi_square)

dof = len(observed) - 1
p_val = 1 - stat.chi2.cdf(chi_square,dof)
print(p_val)

if p_val > significance_val:
    print("Accept null hypothesis.")
else:
    print("Reject null hypothesis.")
```

### Que - 4 : A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import math

p = 60/500
n = 500
z = 1.96

CI_lower = p - z * math.sqrt(p * (1-p) / n)
CI_upper = p + z * math.sqrt(p * (1-p) / n)

print(f"The 95% confidence interval for the true proportion of individuals who smoke is ({CI_lower:.4f}, {CI_upper:.4f})")
```

### Que - 6 : Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the degrees of freedom and create a chi-square distribution
df = 10
x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)
y = chi2.pdf(x, df)

# Plot the chi-square distribution
fig, ax = plt.subplots()
ax.plot(x, y, 'b-', lw=2)

# Shade the area corresponding to a chi-square statistic of 15
xshade = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.975, df), 100)
yshade = chi2.pdf(xshade, df)
ax.fill_between(xshade, yshade, where=(xshade >= 15), color='blue', alpha=0.3)

# Add labels and a title to the plot
ax.set_xlabel('Chi-Square Statistic')
ax.set_ylabel('Probability Density')
ax.set_title('Chi-Square Distribution with 10 Degrees of Freedom')

plt.show()
```

### Que - 7 : A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

To calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke, we can use the following formula:

CI = p̂ ± z*√((p̂(1-p̂))/n)
where:

p̂ is the sample proportion (520/1000 = 0.52)
z is the critical value from the standard normal distribution for a 99% confidence level (z = 2.576)
n is the sample size (1000)

Plugging in the values, we get:

CI = 0.52 ± 2.576*√((0.52(1-0.52))/1000)
CI = 0.52 ± 0.049
CI = (0.471, 0.569)

Therefore, we can be 99% confident that the true proportion of people in the population who prefer Coke is between 0.471 and 0.569.

### Que - 8 : A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05. 

To conduct a chi-square goodness of fit test, we need to follow these steps:

State the null and alternative hypotheses:
Null hypothesis: The observed frequencies match the expected frequencies of a fair coin.
Alternative hypothesis: The observed frequencies do not match the expected frequencies of a fair coin.

Determine the expected frequencies:

For a fair coin, we would expect each outcome (heads and tails) to occur with equal probability, or 0.5.
Therefore, the expected frequency for tails is 0.5 * 100 = 50, and the expected frequency for heads is also 50.

Calculate the test statistic:

The test statistic for a chi-square goodness of fit test is calculated as:
X² = Σ((Oᵢ - Eᵢ)² / Eᵢ)

where:
Oᵢ is the observed frequency for category i
Eᵢ is the expected frequency for category i
Σ is the sum over all categories

Plugging in the values for this problem, we get:
X² = ((45 - 50)² / 50) + ((55 - 50)² / 50)
X² = 1

Determine the critical value:

We have 2 categories (heads and tails) and want to use a significance level of 0.05.
Therefore, the degrees of freedom for this test is 2 - 1 = 1.
Using a chi-square distribution table or calculator, the critical value for a chi-square distribution with 1 degree of freedom and a significance level of 0.05 is 3.84.

Make a decision and interpret the results:

The calculated test statistic (X² = 1) is less than the critical value (3.84).
Therefore, we fail to reject the null hypothesis.
We can conclude that there is not enough evidence to suggest that the observed frequencies do not match the expected frequencies of a fair coin at a significance level of 0.05.

In summary, based on the chi-square goodness of fit test, we cannot reject the null hypothesis that the coin is fair, and therefore we do not have evidence to support the researcher's hypothesis that the coin is biased towards tails.

### Que - 11 : A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

To conduct a hypothesis test for the population mean, we need to follow these steps:

State the null and alternative hypotheses:

Null hypothesis: The population mean is equal to 70 (μ = 70).
Alternative hypothesis: The population mean is significantly different from 70 (μ ≠ 70).

Calculate the test statistic:

The test statistic for a hypothesis test for the population mean is calculated as:
t = (x̄ - μ) / (s / √n)

where:
x̄ is the sample mean
μ is the hypothesized population mean under the null hypothesis
s is the sample standard deviation
n is the sample size

Plugging in the values for this problem, we get:

t = (72 - 70) / (10 / √30)
t = 1.10

Determine the critical value:

We have 29 degrees of freedom (since the sample size is 30).
Using a t-distribution table or calculator, the critical values for a t-distribution with 29 degrees of freedom and a significance level of 0.05 are -2.045 and 2.045 (two-tailed test).

Make a decision and interpret the results:

The calculated test statistic (t = 1.10) falls within the acceptance region between -2.045 and 2.045.
Therefore, we fail to reject the null hypothesis.
We do not have sufficient evidence to conclude that the population mean is significantly different from 70 at a significance level of 0.05.
Note that we cannot make any conclusions about the true population mean from this test alone, but we can say that we did not find evidence to reject the null hypothesis that the population mean is 70.

In summary, based on the hypothesis test for the population mean, we do not have sufficient evidence to suggest that the population mean is significantly different from 70 at a significance level of 0.05.
