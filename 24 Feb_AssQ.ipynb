```python
import pandas as pd
```

```python
df = pd.read_csv("csv/loan_data.csv")
```

```python
df
```

```python
# groupby is used to group the data based on unique values. and after then we can peform any operation on that data like count(),
# sum(), min(), max(), mode().
df.groupby('loan_status').count()
```

```python
# to filter out rows and columns we use iloc based on index location.
df.iloc[[1,2,3],[0,1,2]]
```

```python
# to filter out rows and column we use loc based on named columns.
df.loc[[1,2,3],['id','issue_d','loan_amnt']]
```

```python
df1 = pd.DataFrame({'Name':['Yash','Ganesh','Rohit','Vikas'],'Age':[22,32,12,43],'City':['Ahmedabad','Rajasthan','Kutchh','Rajkot']})
df1
```

```python
df2 = pd.DataFrame({'Name':['Ganesh','Rohit','Vikas'],'Friend':['Yash','Vikas','Ganesh',],'Occupation':['Data Engineer','ML Developer','Sr. Java Developer']})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='inner',left_on = 'Name',right_on='Friend')
# pd.merge(df1,df2,how='inner',left_on = 'Name',right_on='Friend').reindex(columns=['Name_x','Name_y','Friend','Age','City','Occupation'])
```

```python
df3 = pd.DataFrame({'Values':[34,23,45,56,23,34,2]})
```

```python
df3
```

```python
from functools import reduce

def total_sum(data):
    return reduce(lambda x,y : x + y,data)

val_data = list(df3['Values'].values)[:3]
total_sum(val_data)
```
