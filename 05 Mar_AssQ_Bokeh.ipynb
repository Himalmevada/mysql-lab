```python
from bokeh.plotting import figure, output_notebook, show
output_notebook()
```

### Ans - 1

```python
p = figure()
p.circle([1,4,6,5,3],[8,5,6,7,4])
show(p)
```

### Ans - 2

#### A glyph is a vectorized graphical shape or marker that is used to represent your data. A Few Categories of Glyphs: Marker: Shapes like circles, diamonds, squares and triangles. Effective for scatter and bubble charts. Line: Single, step and multi-line shapes.

```python
p = figure()
p.diamond([1,4,6,5,3],[8,5,6,7,4],size=20, color="orange", alpha=0.5)
show(p)
```

```python
p = figure(x_axis_label = 'Numbers',y_axis_label = 'Square Root')
p.line([1,2,3,4,5,6],[2,4,9,16,25,36], color="green", alpha=0.9, legend_label='My Legends')
show(p)
```
