### Ans - 1

#### **GET :**
#### GET is used to request data from a specified resource.
#### Ex - /index/login.php?usenrmae=himal&password=himal2002
#### Get request can be cached.
#### GET requests remain in the browser history.
#### We can bookmark the get request. 
#### We can not pass sensitive information like password or bank details.
#### It has some length restrictions(maximum URL length is 2048 characters).
#### It only allows ASCII characters.
#### It is only used to request data from user to server.

#### **POST :**
#### POST is used to send data to a server to create/update a resource.
#### POST request are never cached.
#### POST request do not remain in the browser history.
#### This request cannot be bookmarked.
#### There are no length restrictions. Binary data is also allowed.
