### Q1. What is the mathematical formula for a linear SVM?

In a Support Vector Machine (SVM) with a linear kernel, the goal is to find a hyperplane that best separates the data into different classes. The decision boundary is a hyperplane defined by:

### The Linear Decision Function:
\[ f(\mathbf{x}) = \mathbf{w} \cdot \mathbf{x} + b \]

Where:
- \( \mathbf{x} \) represents the input features.
- \( \mathbf{w} \) is the weight vector.
- \( b \) is the bias term.
- \( f(\mathbf{x}) \) represents the decision function. If \( f(\mathbf{x}) \geq 0 \), it predicts one class; otherwise, it predicts the other class.

### Mathematical Formulation:

Given a dataset with features \( \mathbf{x}_i \) and corresponding labels \( y_i \) where \( y_i \in \{-1, 1\} \) for binary classification:

1. The optimization problem for a linear SVM aims to maximize the margin between the hyperplane and the support vectors while minimizing the classification error:

\[ \text{Minimize:} \frac{1}{2} ||\mathbf{w}||^2 \]

Subject to the constraints:
\[ y_i(\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \text{ for all } i \]

2. The margin is proportional to \( \frac{1}{||\mathbf{w}||} \), and maximizing the margin is equivalent to minimizing \( ||\mathbf{w}||^2 \).

3. The optimization problem is often solved using Lagrange multipliers to derive the dual form, where the decision function can be expressed in terms of a linear combination of support vectors.

### Solution:

The final decision function for predicting the class of a new sample \( \mathbf{x} \) after training the SVM involves calculating:

\[ f(\mathbf{x}) = \mathbf{w} \cdot \mathbf{x} + b \]

Where \( \mathbf{w} \) and \( b \) are determined during the training process to find the optimal hyperplane that separates the classes while maximizing the margin.

The mathematical formulation aims to find the optimal values of \( \mathbf{w} \) and \( b \) that define this hyperplane, ensuring a clear separation between the classes while minimizing classification errors.

### Q2. What is the objective function of a linear SVM?

In a linear Support Vector Machine (SVM), the objective function aims to maximize the margin between the hyperplane and the support vectors while minimizing the norm of the weight vector. This objective function ensures a wide margin separation between classes while controlling for the complexity of the decision boundary.

### Objective Function:

The objective function for a linear SVM can be expressed as follows:

\[ \text{Minimize:} \frac{1}{2} ||\mathbf{w}||^2 \]

Subject to the constraints:
\[ y_i(\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \text{ for all } i \]

Where:
- \( \mathbf{w} \) is the weight vector.
- \( ||\mathbf{w}||^2 \) represents the squared norm (magnitude) of the weight vector.
- \( \mathbf{x}_i \) represents the input features.
- \( y_i \) represents the corresponding class labels (either -1 or 1 for binary classification).
- \( b \) is the bias term.

### Objective Explanation:

1. **Minimizing Norm of Weight Vector:** The term \( \frac{1}{2} ||\mathbf{w}||^2 \) in the objective function aims to minimize the norm (or length) of the weight vector \( \mathbf{w} \). This encourages the optimization process to find a decision boundary with a small weight vector, promoting a simpler model.

2. **Subject to Constraints:** The constraints \( y_i(\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \) enforce that all training samples are correctly classified and lie outside the margin (i.e., the correct class is on the correct side of the decision boundary with a margin of at least 1).

### Objective Interpretation:

The objective of the linear SVM is to find the optimal hyperplane (defined by \( \mathbf{w} \) and \( b \)) that separates the classes while maximizing the margin. Maximizing the margin ensures better generalization and robustness of the model by maximizing the separation between classes and reducing the risk of overfitting to the training data.

### Q3. What is the kernel trick in SVM?

The kernel trick is a method used in Support Vector Machines (SVMs) to handle nonlinear relationships between features without explicitly transforming the data into a higher-dimensional space. It allows SVMs to learn and model complex, nonlinear decision boundaries by implicitly operating in a higher-dimensional space while still working in the original feature space.

### Key Points:

1. **Nonlinear Transformations:** SVMs with a linear kernel work well for linearly separable data. However, when data isn't linearly separable, nonlinear transformations can help. These transformations map the original feature space into a higher-dimensional space, where the data might become linearly separable.

2. **Computational Complexity:** Directly computing transformations to higher-dimensional spaces can be computationally expensive, especially with many features or complex transformations.

3. **The Kernel Trick:** Instead of explicitly transforming the data, the kernel trick computes the inner product (dot product) between the transformed data points in the higher-dimensional space without explicitly performing the transformation. This allows SVMs to implicitly operate in the higher-dimensional space while staying in the original feature space.

4. **Kernel Functions:** Kernel functions represent different transformations, allowing SVMs to work in higher-dimensional spaces without explicitly performing the transformations. Popular kernels include:
   - **Linear Kernel:** \( K(\mathbf{x}_i, \mathbf{x}_j) = \mathbf{x}_i \cdot \mathbf{x}_j \) (works in the original feature space).
   - **Polynomial Kernel:** \( K(\mathbf{x}_i, \mathbf{x}_j) = (\gamma \mathbf{x}_i \cdot \mathbf{x}_j + r)^d \) (maps to polynomial feature space).
   - **Gaussian RBF Kernel:** \( K(\mathbf{x}_i, \mathbf{x}_j) = \exp(-\gamma ||\mathbf{x}_i - \mathbf{x}_j||^2) \) (maps to infinite-dimensional space).

### Benefits of the Kernel Trick:

1. **Nonlinear Separability:** It allows SVMs to handle nonlinear relationships between features without explicitly transforming the data into higher-dimensional spaces.

2. **Computational Efficiency:** It avoids the computational burden of explicitly working in higher-dimensional spaces, as it computes inner products directly in the original space.

3. **Versatility:** Different kernel functions provide flexibility to model various types of relationships, enabling SVMs to handle diverse datasets effectively.

The kernel trick significantly enhances the flexibility and power of SVMs, enabling them to effectively learn complex decision boundaries without the need for explicit feature space transformations.

### Q4. What is the role of support vectors in SVM Explain with example 

Support vectors play a crucial role in Support Vector Machines (SVMs) as they define the decision boundary and influence the model's performance. They are the data points that lie closest to the decision boundary and are instrumental in determining the optimal hyperplane that separates different classes. Let's illustrate their role with an example:

### Example:

Consider a 2D classification problem where we want to separate two classes, represented by red and blue points:



In this case:
- Red points represent one class (label 1).
- Blue points represent another class (label -1).

### Role of Support Vectors:

1. **Defining the Decision Boundary:** Support vectors are the data points closest to the decision boundary, which is determined by the hyperplane.
   
   - In the example, the support vectors are the highlighted points (both red and blue) nearest to the decision boundary, shown as the dashed line.

2. **Determining the Margin:** The margin is the distance between the decision boundary and the closest support vectors.
   
   - The widest possible margin is achieved when the decision boundary is equidistant from the nearest support vectors, as depicted in the example.

3. **Model's Dependence on Support Vectors:** In many cases, only a small subset of data points (the support vectors) affects the position and orientation of the decision boundary. This characteristic allows SVMs to be memory-efficient and less sensitive to outliers or irrelevant data points.

### SVM Training Process:

During the training process of an SVM:
- The model identifies the support vectors.
- These support vectors contribute to the formulation of the decision boundary.
- The goal is to find the hyperplane that maximizes the margin while minimizing classification errors, with support vectors defining this boundary.

### Importance of Support Vectors:

- SVMs primarily rely on support vectors to generalize and make predictions.
- They influence the construction of the decision boundary and, by extension, the model's predictive capacity.
- Changes in the position or characteristics of support vectors can alter the decision boundary and subsequently affect the model's predictions.

### Summary:

Support vectors are critical in SVMs as they dictate the decision boundary's position and margin, providing the foundation for the model's classification capability. Their role ensures that the SVM focuses on the most informative data points to construct an optimal hyperplane that separates different classes while maximizing the margin.

### Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?

```python
# 1 : Hyperplane
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# Generate a simple dataset with two classes
X, y = make_blobs(n_samples=100, n_features=2, centers=2, cluster_std=1.0, random_state=42)

# Fit an SVM model with a linear kernel
svm = SVC(kernel='linear')
svm.fit(X, y)

# Scatter plot of the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', label='Data Points')

# Get the coefficients of the hyperplane
w = svm.coef_[0]
b = svm.intercept_[0]

# Plotting the hyperplane (decision boundary)
x_plot = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
y_plot = (-w[0] * x_plot - b) / w[1]
plt.plot(x_plot, y_plot, color='black', linestyle='-', label='Hyperplane')

plt.title('Hyperplane in SVM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# 2 : Margin

# Scatter plot of the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', label='Data Points')

# Plotting the hyperplane
plt.plot(x_plot, y_plot, color='black', linestyle='-', label='Hyperplane')

# Plotting the margins (support vectors)
support_vectors = svm.support_vectors_
for i in range(len(support_vectors)):
    plt.scatter(support_vectors[i][0], support_vectors[i][1], s=200, facecolors='none', edgecolors='k', label='Support Vectors' if i == 0 else '')

# Calculate the distances to the hyperplane for plotting margins
distances = svm.decision_function(X)
for i, d in enumerate(distances):
    if np.abs(d) < 1:
        plt.scatter(X[i][0], X[i][1], s=100, facecolors='none', edgecolors='r', label='Marginal Plane' if i == 0 else '')

plt.title('Margin in SVM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

### 3. Soft Margin vs. Hard Margin:
#### Hard Margin:
- In a hard margin SVM, the margin is maximized without allowing any data points to violate it.
- If the data is not linearly separable, a hard margin SVM won't find a feasible solution.
#### Soft Margin:
- A soft margin SVM allows some data points to be within the margin or even on the wrong side of the margin.
- It introduces a penalty (C parameter) for misclassified points, allowing for a trade-off between maximizing the margin and minimizing classification errors.

### Q6. SVM Implementation through Iris dataset.
- Load the iris dataset from the scikit-learn library and split it into a training set and a testing set.
- Train a linear SVM classifier on the training set and predict the labels for the testing set.
- Compute the accuracy of the model on the testing set.
- Plot the decision boundaries of the trained model using two of the features.
- Try different values of the regularisation parameter C and see how it affects the performance of the model.

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import matplotlib.pyplot as plt
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
df = pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
df
```

```python
df["target"] = dataset.target
```

```python
df
```

```python
X = df.iloc[:,:-1]
y = df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
svc = SVC(kernel="linear")
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
svc = SVC(C=0.3)
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
svc = SVC(C=100)
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
