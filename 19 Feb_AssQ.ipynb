### Ans - 3

#### App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler.
#### Example: In our application, the URL(/) is associated with the root URL.

### Ans - 4

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return """Company Name: ABC Corporation<br>
            Location: India<br>
            Contact Detail: 999-999-9999"""

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```

### Ans - 5

#### To build a URL to a specific function, use the url_for() function. It accepts the name of the function as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.

```python
from flask import Flask
from flask import url_for

app = Flask(__name__)

@app.route("/")
def home():
    return "This is my home page."


@app.route("/aboutus")
def about():
    return "This is my about us page."


@app.route('/user/<category>/<page_no>')
def post(category,page_no):
    return f'Category : {category} and Page no : {page_no}'


with app.test_request_context():
    print(url_for('home'))
    print(url_for('about', next='/'))
    print(url_for('post', category='Technology',page_no=5))

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
